```python
#importing important modules to download and extract data set into csv file.
import os
import tarfile
from six.moves import urllib
```

```python
#downloading dataset and storing in dataset/housing/housing.tgz
DOWNLOAD_ROOT="https://raw.githubusercontent.com/ageron/handson-ml2/master/"
HOUSING_PATH= os.path.join("datasets","housing")
HOUSING_URL=DOWNLOAD_ROOT + "datasets/housing/housing.tgz"
```

```python
#declaring function for extracting and fetching dataset
def fetch_housing_data(housing_url=HOUSING_URL,housing_path=HOUSING_PATH):
    if not os.path.isdir(housing_path):
        os.makedirs(housing_path)
    tgz_path=os.path.join(housing_path,"housing.tgz")
    urllib.request.urlretrieve(housing_url,tgz_path)
    housing_tgz=tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()
```

```python
#calling above declared function
fetch_housing_data()
```

```python
#importing pandas to do csv operation.
import pandas as pd
```

```python
#declaring function to read csv data
def load_housing_data(housing_path=HOUSING_PATH):
    csv_path=os.path.join(housing_path,"housing.csv")
    return pd.read_csv(csv_path)
```

```python
#now, creating a test set using iloc(when indecies are not known) and random permutation
import numpy as np
def test_train_split(data,test_ratio):
    shuffeled_incdices=np.random.permutation(len(data))
    test_set_size = int(len(data)*test_ratio)
    test_indices=shuffeled_incdices[:test_set_size]
    train_indices=shuffeled_incdices[test_set_size:]
    return data.iloc[train_indices],data.iloc[test_indices]
```

# Custom Transformer

before going for pipeline and feature scaling we should first have a look at custom transformer because skit-learn use duck typing instead of inheritance, so it's better we create our own class and set of function, because we can use TransformerMixin and BaseEstimator we can get two extra methods set_params() and get_params() and is useful for auto hyper-parameter tuning.

```python
from sklearn.base import TransformerMixin , BaseEstimator
rooms_ix, bedrooms_ix, population_ix, households_ix = 3, 4, 5, 6
class combinedattributesadder(BaseEstimator,TransformerMixin):
    
    def __init__(self,add_bedrooms_per_room = True):#hyperparameter
        self.add_bedrooms_per_room=add_bedrooms_per_room
    def fit(self,x,y=None):
        return self #nothing to do
    def transform(self,x,y=None):
        rooms_per_household = x[:,rooms_ix]/x[:,households_ix]
        population_per_household=x[:,population_ix]/x[:,households_ix]
        if self.add_bedrooms_per_room:
            bedroom_per_room = x[:,bedrooms_ix]/x[:,rooms_ix]
            return np.c_[x,rooms_per_household,population_per_household,bedroom_per_room]
        else:
            return np.c_[x,rooms_per_household,population_per_household]
attr_addr=combinedattributesadder(add_bedrooms_per_room=False)
housing_extra_attr=attr_addr.transform(housing.values)
print(housing_extra_attr)
```

# Feature Scaling

Min-Max Scaling(Normalization): Values are shifted and re-scalled between 0 and 1, we substract min value and then divide by max - min, skitlearn has a library or module knwon as MinMaxScaler

Standardization: First mean is substracted (so mean is 0) and then divide by standard deviation (so resulting distribution has unit variance), it is good for outliers, skitlearn has a library or module known as StandardScaler.

# Tranformation Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([('imputer',SimpleImputer(strategy="median")),('attr_addr',combinedattributesadder()),
                        ('std_scaler',StandardScaler())])
housing_num_tr=num_pipeline.fit_transform(housing_num)
print(housing_num_tr)
```

```python
#Instead of applying transforming on numerical and categorical data seprately, now we will use a single skitlearn
#library known as ColumnTransformer
from sklearn.compose import ColumnTransformer

num_attributes=list(housing_num)
cat_attributes=["ocean_proximity"]
full_pipeline = ColumnTransformer([('num',num_pipeline,num_attributes),('cat',OneHotEncoder(),cat_attributes)])
housing_final=full_pipeline.fit_transform(housing)
print(housing_final)
```

## Randomized Search

The grid search approach is fine when you are exploring relatively few combinations, like in the previous example, but when the hyperparameter search space is large, it is often preferable to use RandomizedSearchCV instead. This class can be used in much the same way as the GridSearchCV class, but instead of trying out all possible combi nations, it evaluates a given number of random combinations by selecting a random value for each hyperparameter at every iteration. This approach has two main bene fits:

 If you let the randomized search run for, say, 1,000 iterations, this approach will explore 1,000 different values for each hyperparameter (instead of just a few val ues per hyperparameter with the grid search approach).
 You have more control over the computing budget you want to allocate to hyper parameter search, simply by setting the number of iterations.
